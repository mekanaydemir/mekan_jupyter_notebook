```python
"""
Program
1'den 1000'e kadar olan asal sayıları ekrana yazdıran bir program yazın. 
Daha sonra bir tane decorator fonksiyon kullanarak bu fonksiyona 1'den 1000'e kadar olan mükemmel sayıları 
yazdırma özelliği ekleyin.
"""
```

```python
def ciftleri_yazdir(func):
    def mekan(sayilar):
        for i in sayilar:
            if i%2==0:
                print(i," cift sayidir")
        func(sayilar)
    return mekan
```

```python
@ciftleri_yazdir
def yazdir(sayi):
    for i in sayi:
        print(i)
```

```python
yazdir(range(1,101))
```
